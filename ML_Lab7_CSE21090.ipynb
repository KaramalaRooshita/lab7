```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("NID_data.csv")
data
```

# A1

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import StandardScaler as ss
from sklearn.svm import SVC

class1 = "normal"
class2 = "anomaly"
selected_data = data[(data['class'] == class1) | (data['class'] == class2)]
selected_data = pd.get_dummies(selected_data, columns=['protocol_type', 'service', 'flag'])

X = selected_data.drop(columns=['class'])
y = selected_data['class']
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.3, random_state=42)

scaler = ss()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

svm_model = SVC(kernel='linear', C=1.0)  
svm_model.fit(X_train, y_train)

accuracy = svm_model.score(X_test, y_test)
print(f"Accuracy: {accuracy}")
```

```python
from sklearn import svm

clf = svm.SVC()
clf.fit(X_train, y_train)
support_vector_indices = clf.support_

support_vectors = X_train[support_vector_indices]

print("Support vectors :", support_vectors)
```

# A2

```python
from sklearn import svm
accuracy = clf.score(X_test, y_test)

print(f"Accuracy: {accuracy}")
```

# A3

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear', C=1)
svm.fit(X_train, y_train)
predicted_labels = svm.predict(X_test)
decision_values = svm.decision_function(X_test)

accuracy = (predicted_labels == y_test).mean()

print("Predicted Labels:", predicted_labels)
print("\nTrue Labels:", y_test)
print("\nDecision Function Values:", decision_values)
print(f"\nAccuracy: {accuracy * 100:.2f}%")
```

# A4

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

kernel_functions = ['linear', 'poly', 'rbf', 'sigmoid']

for kernel_func in kernel_functions:
    svm_classifier = SVC(kernel=kernel_func)
    svm_classifier.fit(X_train, y_train)
    y_pred_svm = svm_classifier.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred_svm)

    print(f"Kernel: {kernel_func}, Accuracy: {accuracy:.2f}")
```
